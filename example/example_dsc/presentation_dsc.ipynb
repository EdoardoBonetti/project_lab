# Read the data and plot it

As a good practice one needs to just read the data and plot it, to do so we use the simple code below.

```python

input_dir = "/home/ebonetti/Desktop/package_lab/project_lab/example/example_dsc/input_data/2perc"
input_name = "231116_RBX_RSO_2perc_A_1st_heating_raw_data.txt"
dsc = SingleDSC(file_name=input_name, file_dir= input_dir)
print(dsc.data)

# PLOT THE DATA
output_dir = "output_data/single"
dsc.data_plotter(output_dir=output_dir, name="plot_single.png")
```

We want to truncate the temperature since it does wierd things at the end of the curve and at the beginning.

Moreover we can see that the data is not smooth, to adjust it we give a smoothing factor. 


```python

input_dir = "/home/ebonetti/Desktop/package_lab/project_lab/example/example_dsc/input_data/2perc"
input_name = "231116_RBX_RSO_2perc_A_1st_heating_raw_data.txt"
dsc = SingleDSC(file_name=input_name, file_dir=input_dir)
print(dsc.data)

# CLEAN THE DATA AND SAVE IT
kwargs_cleaner = {
    "min_temp":25, "max_temp":85,
    "p": 0.05, "lam" : 5e12, "niter" : 100}
dsc.data_check(**kwargs_cleaner)



```

```python

dsc.data_cleaner(**kwargs_cleaner)

# save the cleaned data
output_dir = "/home/ebonetti/Desktop/package_lab/project_lab/example/example_dsc/output_data/2perc"
dsc.data_saver(output_dir=output_dir, name="cleaned_data.txt")
print(dsc.data)

# PLOT THE DATA
dsc.data_plotter(output_dir=output_dir, name="plot_single.png", save=False)

```

One can create a loop to check on all the data in a folder:

```python
kwargs_cleaner = {
    "min_temp": 25, "max_temp": 85,
    "p": 0.05, "lam": 5e12, "niter": 100}

input_dir = "/home/ebonetti/Desktop/package_lab/project_lab/example/example_dsc/input_data/2perc"

for dir_ in os.listdir(input_dir):
    dsc = SingleDSC(file_name=dir_, file_dir=input_dir)
    dsc.data_check(**kwargs_cleaner)
    
```
